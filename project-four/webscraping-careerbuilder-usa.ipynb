```python
import requests
from bs4 import BeautifulSoup
from time import sleep
import numpy as np
import re
import os
from selenium import webdriver
import pandas as pd

def get_job_urls(driver, job="data-scientist"):
    job_urls = []
    for page_num in range(1, 13):
        driver.get("https://www.careerbuilder.com/jobs-{}-in-usa?page_number={}".format(job, page_num))
        
        sleep(3 * np.random.uniform())
        
        html = BeautifulSoup(driver.page_source, 'lxml')

        for url in html.find_all(name='h2', attrs={'class':'job-title show-for-medium-up'}):
            job_urls.append('https://www.careerbuilder.com{}'.format(url.a['href']))
    return job_urls

def get_job_info(url, driver, job_dict):    
    driver.get(url)
    
    sleep(3 * np.random.uniform())
    
    soup = BeautifulSoup(driver.page_source, 'lxml')
    
    # Get the job title:
    try:
        job_title = soup.find('div', attrs={'class':'small-12 item'}).h1.text.replace('\n','')
        job_dict["title"].append(job_title)
    except AttributeError:
        job_dict["title"].append('NA')
    
    # Get the industry of the company, e.g. Medical:
    try:
        job_industry = soup.find('div', {'class': 'tag', 'id': 'job-industry'}).text.replace('\n','')
        job_dict["industry"].append(job_industry)
    except AttributeError:
        job_dict["industry"].append('NA')
    
    # Get the nature/category of the job, e.g. IT:
    try:
        job_category = soup.find('div', {'class': 'tag', 'id': 'job-categories'}).text.replace('\n','')
        job_dict["category"].append(job_category)
    except AttributeError:
        job_dict["category"].append('NA')
    
    # Get the name of the company:
    try:
        company_name = soup.find('div', {'class': 'company-info'}).header.h4.text.replace('\n','')
        job_dict["company"].append(company_name)
    except AttributeError:
        job_dict["company"].append('NA')
    
    # Get the location of the job, e.g. California
    try:
        job_location = soup.find('ul', {'class': 'address'}).li.text.replace('\n','')
        job_dict["location"].append(job_location)
    except AttributeError:
        job_dict["location"].append('NA')
    
    # Get job description and requirements
    job_description = soup.find_all(name='div', attrs={'class':'description'})
    description = ''
    for i in job_description:
        description += i.text
    job_dict["details"].append(description)

    # Get data that appears as "Job Snapshot" tags:
    job_snapshot = soup.find_all('div', {'class': 'tag'})

    salary_count = False
    experience_count = False
    employment_type_count = False
    degree_count = False
    travel_count = False
    
    for i in job_snapshot:
        if '$' in i.text and salary_count == False:
            # Get job salary:
            salary_count = True
            job_dict["salary"].append(i.text.replace('\n',''))
        elif 'contract' in i.text.lower() or 'time' in i.text.lower() and employment_type_count == False:
            # Get employment_type:
            employment_type_count = True
            job_dict["employment_type"].append(i.text.replace('\n',''))
        elif 'degree' in i.text.lower() and degree_count == False:
            # Get degree:
            degree_count = True
            job_dict["degree"].append(i.text.replace('\n',''))
        elif 'travel' in i.text.lower() and travel_count == False:
            # Get travel:
            travel_count = True
            job_dict["travel"].append(i.text.replace('\n',''))
        elif 'experience' in i.text.lower() and experience_count == False:
            # Get experience:
            experience_count = True
            job_dict["experience"].append(i.text.replace('\n',''))
            
    if salary_count == False:
        job_dict["salary"].append('NA') 
    if employment_type_count == False:
        job_dict["employment_type"].append('NA') 
    if degree_count == False:
        job_dict["degree"].append('NA') 
    if travel_count == False:
        job_dict["travel"].append('NA')
    if experience_count == False:
        try:
            exp = re.search(r'[1-9]?(-| - | to )*[1-9]\+? [Yy]ears?', description).group(0)
            job_dict["experience"].append(exp)
        except AttributeError:
            job_dict["experience"].append('NA')
        
    return 

def scrape_jobs_to_csv(job="data-scientist"):
    chromedriver = "/Users/jasminetan/code/jt-dsi-exercises/chromedriver/chromedriver"
    os.environ["webdriver.chrome.driver"] = chromedriver
    
    # Create driver:
    driver = webdriver.Chrome(executable_path=chromedriver)

    # Gather urls from which to extract job information:
    job_urls = get_job_urls(driver, job=job)
        
    # Check 300 jobs per job type:
    print("Number of jobs: {}".format(len(job_urls)))
    
    # Initiate lists for various column headers/features:
    job_dict = {
        'title': [],
        'industry': [],
        'category': [],
        'company': [],
        'location': [],
        'details': [],
        'salary': [],
        'experience': [],
        'employment_type': [],
        'degree': [],
        'travel': []
               }
    
    # Get job info for each url collected
    for url in job_urls:
        print job_urls.index(url) + 1
        get_job_info(url, driver, job_dict)
    

    # Add columns to a dataframe:
    job_df = pd.DataFrame(job_dict)
    
    job_df.to_csv('{}.csv'.format(job), encoding='utf-8')
    driver.close()

scrape_jobs_to_csv(job="data-scientist")
scrape_jobs_to_csv(job="data-analyst")
scrape_jobs_to_csv(job="data-engineer")
scrape_jobs_to_csv(job="data-architect")
scrape_jobs_to_csv(job="business-analyst")
scrape_jobs_to_csv(job="business-intelligence")
scrape_jobs_to_csv(job="machine-learning-engineer")
scrape_jobs_to_csv(job="machine-learning-scientist")
scrape_jobs_to_csv(job="artificial-intelligence-engineer")
scrape_jobs_to_csv(job="artificial-intelligence-scientist")
scrape_jobs_to_csv(job="database-developer")
scrape_jobs_to_csv(job="database-administrator")
scrape_jobs_to_csv(job="statistician")
scrape_jobs_to_csv(job="statistical-programmer")
```
